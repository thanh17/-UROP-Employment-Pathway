```python
def get_data(connection, code,t,scale):
    """
    Given a database connection, onetsoc_code, t: (skills/abilities/knowledge),
    scale: level/importance
    return df that contains:
        job title    element_name   scale_name   data_value   
    """
    if scale == 'both':
        df = pd.read_sql('''
        SELECT title, element_name, element_id, scale_name, scale_id, data_value
        FROM employment.''' + t + '''
        LEFT JOIN employment.occupation_data USING (onetsoc_code)
        LEFT JOIN employment.content_model_reference USING (element_id)
        LEFT JOIN employment.scales_reference USING (scale_id)
        WHERE onetsoc_code LIKE \'''' + code + '''\' 
        ORDER BY data_value DESC;''', con = connection)
    else:
        df = pd.read_sql('''
        SELECT title, element_name, element_id, scale_name, scale_id, data_value
        FROM employment.''' + t + '''
        LEFT JOIN employment.occupation_data USING (onetsoc_code)
        LEFT JOIN employment.content_model_reference USING (element_id)
        LEFT JOIN employment.scales_reference USING (scale_id)
        WHERE scale_id LIKE \'''' + scale + '''\' 
        AND onetsoc_code LIKE \'''' + code + '''\' 
        ORDER BY data_value DESC;''', con = connection)
    return df
```

```python
df_test = get_data(connection,'17-2011.00','skills','both')
# df_test[df_test['element_id'].str.contains('2.A.1')]
df_test
```

```python
def type_values(df,list_val,name_dict):
    '''
    Create df with 
    title    element_name     avg_value  
    '''
    
    data = {'title': [], 'element_name': [], 'avg_value': []}
    
    for val in list_val:
        data['title'].append(df['title'].any())
        df_val = df[df['element_id'].str.contains(val)]
        data['element_name'].append(name_dict[val])
        data['avg_value'].append(df_val['data_value'].mean())
    return pd.DataFrame.from_dict(data = data)
```
