# Abilities/Skills/Knowledge for:
## Aerospace Engineers, Agricultural Engineers, and Biomedical Engineers

```python
def get_data(connection, code,t,scale):
    """
    Given a database connection, onetsoc_code, t: (skills/abilities/knowledge),
    scale: level/importance
    return df that contains:
        job title    element_name   scale_name   data_value   
    """
    if scale = 'both':
        df = pd.read_sql('''
        SELECT title, element_name, scale_name, data_value
        FROM employment.''' + t + '''
        LEFT JOIN employment.occupation_data USING (onetsoc_code)
        LEFT JOIN employment.content_model_reference USING (element_id)
        LEFT JOIN employment.scales_reference USING (scale_id)
        WHERE scale_name LIKE \'''' + scale + '''\' 
        ORDER BY data_value DESC;''', con = connection)
    else:
        df = pd.read_sql('''
        SELECT title, element_name, scale_name, data_value
        FROM employment.''' + t + '''
        LEFT JOIN employment.occupation_data USING (onetsoc_code)
        LEFT JOIN employment.content_model_reference USING (element_id)
        LEFT JOIN employment.scales_reference USING (scale_id)
        WHERE scale_name LIKE \'''' + scale + '''\' 
        AND onetsoc_code LIKE \'''' + code + '''\' 
        ORDER BY data_value DESC;''', con = connection)
    return df
```

### Aerospace engineers

```python
# Aero = 17-2011.00
# Agri = 17-2021.00
# biom = 17-2031.00
df_aero_skills = get_data(connection,'17-2011.00','skills','importance')
df_aero_skills

# df_aero_abilities = pd.read_sql('''
# SELECT title, element_name, scale_name, data_value
# FROM employment.abilities 
# LEFT JOIN employment.occupation_data USING (onetsoc_code)
# LEFT JOIN employment.content_model_reference USING (element_id)
# LEFT JOIN employment.scales_reference USING (scale_id)
# WHERE scale_name LIKE 'importance'
# AND onetsoc_code LIKE '17-2011.00'
# ORDER BY data_value DESC;''', con = connection)

# df_aero_knowledge = pd.read_sql('''
# SELECT title, element_name, scale_name, data_value
# FROM employment.knowledge 
# LEFT JOIN employment.occupation_data USING (onetsoc_code)
# LEFT JOIN employment.content_model_reference USING (element_id)
# LEFT JOIN employment.scales_reference USING (scale_id)
# WHERE scale_name LIKE 'importance'
# AND onetsoc_code LIKE '17-2011.00'
# ORDER BY data_value DESC;''', con = connection)
```
